# bivariate analysis

## 1. outputs vs age

```python
bins=[20,30,40,50,60,70,80] # bin range

data['extra_col']=pd.cut(data['age'],bins)
age=pd.crosstab(data['extra_col'],data['output']) 
age.div(age.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True) 

plt.xlabel('age') 
P = plt.ylabel('output')
```

## Following inferences can be made from the above bar plots:

1. People having age 40-60 are more likely to have heart disease whereas people having higher heart rate are very much probable to suffer from heart attack.

## 2. outputs vs sex

```python
sex=pd.crosstab(data['sex'],data['output'])
sex.div(sex.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(4,4))
plt.show() 
```

## Outlier Treatment
As we saw earlier in univariate analysis, chol contains outliers so we have to treat them as the presence of outliers affects the distribution of the data.

```python
Q1 = data.quantile(0.25).loc['chol']
Q3 = data.quantile(0.75).loc['chol']
IQR = Q3 - Q1
print(IQR,data.shape)
```

```python
data = data[~((data.chol < (Q1 - 1.5 * IQR)) | (data.chol > (Q3 + 1.5 * IQR)))]
data.shape
```
